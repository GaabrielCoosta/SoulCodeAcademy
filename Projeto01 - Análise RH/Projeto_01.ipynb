# Estatística Descritiva e Dispersão
# Medidas descritivas
- A estatística descritiva organiza e descreve os dados de três maneiras: por meio de tabelas, de gráficos e de medidas descritivas.

# Medidas de dispersão
- A medidas de dispersão nos ajudam a entender como os dados variam em torno da média, ou seja, também ajudam a descrever a tendência nos dados.

Descrições das variáveis:

1.   Medidas de Tendência Central

  *   md = Média
  *   mediana = Mediana
  *   moda = Moda

2.   Medidas de Dispersão

  *   dp = Desvio Padrão
  *   va = Variância
  *   q1 = Quartil 1
  *   q3 = Quartil 3
  *   min = Mínimo valor
  *   max = Máximo valor
  *   amp = Amplitude Total
  *   aiq = Amplitude Interquatilica




```python
# carregando os dados da tabela filtrada dos funciónarios que pediram demissão

desgastados = pd.read_csv("/content/drive/MyDrive/Projeto01-Colab_GitHub/Baserh_Desgastados  - basedesgaste (1).csv")
#display(desgastados)


# *********** Medidas de Tendencia Central ************

# Média
md = desgastados.mean()
#print("A média dos atributos são:")
#display(md)

# Mediana
mediana = desgastados.median()
#print("A mediana dos atributos são:")
#display(mediana)

# Moda
moda = desgastados.mode()
#print("A moda dos atributos são:")
#display(moda)


# *********** Medidass de dispersão *************
# Máximo
#print("O máximo dos atributos são:")
max = desgastados.max()
#display(max)

# Mínimo
min = desgastados.min()
#print("O mínimo dos atributos são:")
#display(min)

# Desvio Padrão
dp = desgastados.std()
#print("O desvio padrão dos atributos são:")
#display(dp)

# Variância
va = desgastados.var()
#print("A variância dos atributos são:")
#display(va)

# Amplitude Total
amp = desgastados.max() - desgastados.min()
#print("A amplitude total dos atributos são:")
#display(amp)

# Quartil 1
q1 = desgastados.quantile(.025)
#print("O quartil 1 dos atributos são:")
#display(md)

# Quartil 3
q3 = desgastados.quantile(0.75)
#print("O quartil 3 dos atributos são:")
#display(md)

# Amplitude Interquartil
aiq = desgastados.quantile(.75) - desgastados.quantile(.25)
#print("A amplitude interqualil dos atributos:")
#display(aiq)

```
