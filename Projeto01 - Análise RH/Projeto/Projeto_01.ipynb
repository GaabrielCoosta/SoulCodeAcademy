# Descriptive Statistics and Dispersion
### Descriptive measures

### Dispersion measures
- *Scattering measures help us understand how the data varies around the mean, i.e. they also help describe the trend in the data*







#1. Measures of Central Tendency

**Variable Descriptions:**

  *   *md = Média*
  *   *mediana = Mediana*
  *   moda = Moda

```python
import pandas as pd
# carregando os dados da tabela filtrada dos funciónarios que pediram demissão

basedesgaste = pd.read_csv('/content/drive/MyDrive/Projeto01-Colab_GitHub/desgastebase.csv', sep=',', engine='python')
#display(desgastados)

# Média
#print("-------- Média ---------")
md = basedesgaste.mean()
pd.set_option('display.float_format','{:.2f}'.format)
#display(md)


# Mediana
#print("-------- Mediana ---------")
mediana = basedesgaste.median()
pd.set_option('display.float_format','{:.2f}'.format)
#display(mediana)

# Moda
print("---------- Moda -----------")
moda = basedesgaste.mode()
moda = moda.dropna(how='all')
moda = moda.fillna('Unimodal')
#display(moda[0:2])

```

#2. Dispersion Measures

**Variable descriptions:**

  *   *dp = Desvio Padrão*
  *   *va = Variância*
  *   *q1 = Quartil 1*
  *   *q3 = Quartil 3*
  *   *min = Mínimo valor*
  *   *max = Máximo valor*
  *   *amp = Amplitude Total*
  *   aiq = Amplitude Interquatilica


```python
import pandas as pd

# carregando os dados da tabela filtrada dos funciónarios que pediram demissão
basedesgaste = pd.read_csv('/content/drive/MyDrive/Projeto01-Colab_GitHub/desgastebase.csv', sep=',', engine='python')
#display(desgastados)

# Máximo
#print("O máximo dos atributos são:")
max = basedesgaste.max()
#display(max)

# Mínimo
min = basedesgaste.min()
#print("O mínimo dos atributos são:")
#display(min)

# Desvio Padrão
dp = basedesgaste.std()
#print("O desvio padrão dos atributos são:")
#display(dp)

# Variância
va = basedesgaste.var()
#print("A variância dos atributos são:")
#display(va)

# Amplitude Total
amp = basedesgaste.max() - basedesgaste.min()
#print("A amplitude total dos atributos são:")
#display(amp)

# Quartil 1
q1 = basedesgaste.quantile(.025)
#print("O quartil 1 dos atributos são:")
#display(md)

# Quartil 3
q3 = basedesgaste.quantile(0.75)
#print("O quartil 3 dos atributos são:")
#display(md)

# Amplitude Interquartil
aiq = basedesgaste.quantile(.75) - basedesgaste.quantile(.25)
#print("A amplitude interqualil dos atributos:")
#display(aiq)

```
