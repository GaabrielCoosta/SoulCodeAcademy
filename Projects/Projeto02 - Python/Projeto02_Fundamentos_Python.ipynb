## Arithmetic average
- The arithmetic mean is a measure of central tendency that represents the average value of a set of numbers. It is calculated by adding all the numbers and dividing the result by the total amount of numbers present.

- The arithmetic mean is calculated as follows:

Let x1, x2, x3, ..., xn be the numbers in the set, the arithmetic mean (denoted by μ or x̄) is given by:

**μ = (x1 + x2 + x3 + ... + xn) / n ***

**Where:**

x1, x2, x3, ..., xn are the numbers in the set.
n is the total number of elements in the set.

Media-Aritmetica.jpg

### Function created for Arithmetic Mean

```python
def media_aritmetica(lista1):
    soma_total = sum(lista1) # the sum() function, adds all the elements of the list
    total_elementos = len(lista1) # the len() function, informs how many elements there are in the list
    media = soma_total / total_elementos # Divide the total sum of the list by the number of elements
    return media
```

## Median
- The median is a statistical measure that represents the central value of an ordered data set. In other words, it is the value that divides the dataset into two equal parts, where half of the values ​​are below the median and the other half are above it.




### To calculate the median, follow these steps:

* Sort data in ascending order
* If the total number of elements is odd, the median is the middle value
* If the total number of elements is even, the median is the average of the two middle values

mediana.png

### Function created to calculate the Median

```python
def mediana(lista):
    ordena_lista = sorted(lista) # sorted() sorts the list in ascending order
    tamanho = len(ordena_lista)

    if tamanho % 2 == 1:
        # If the list has an odd number of elements
        mediana = ordena_lista[tamanho // 2]
    else:
        # If the list has an even number of elements
        indice1 = tamanho // 2 - 1
        indice2 = tamanho // 2
        mediana = (ordena_lista[indice1] + ordena_lista[indice2]) / 2
    return mediana
```

##Sample Variance
- The sample variance is a statistical measure that measures the dispersion of data in relation to the mean in a sample. It provides an estimate of the variability or dispersion of individual values ​​relative to the sample mean


- The formula for calculating the sample variance is as follows:

formula-variancia.jpg

- s^2 represents the sample variance;
- Σ denotes the sum of the elements;
- x is each element in the sample;
- x̄ is the sample mean;
- n is the sample size;

In summary, sample variance is calculated by summing the squared differences between each element in the sample and the sample mean, divided by the sample size minus one.

Sample variance is useful for measuring data dispersion and is widely used in statistics. However, it's important to note that sample variance can be sensitive to outliers in the data, which can affect its interpretation.

### Function created to calculate Sample Variance

```python
def variancia_amostral(lista):
    n = len(lista) #
    media = sum(lista) / n
    soma_dif_quadrados = sum((x - media) ** 2 for x in lista)
    variancia_amostral = soma_dif_quadrados / (n - 1)
    return variancia_amostral
```

##Sample Standard Deviation
- A statistical measure that quantifies the spread of values ​​relative to the mean in a sample. It is the square root of the sample variance and provides an estimate of the dispersion of individual values ​​in relation to the sample mean

- The formula for calculating the sample standard deviation is as follows:

desviopadrao.png

- s represents the sample standard deviation;
- Σ denotes the sum of the elements;
- x is each element in the sample;
- x̄ is the sample mean;
- n is the sample size;

In summary, the sample standard deviation is calculated by summing the square of the difference between each sample element and the sample mean, dividing by the sample size minus one, and then taking the square root of the result.

The sample standard deviation is a measure commonly used to quantify data dispersion in relation to the mean. It provides a measure of variability that is expressed in the same unit as the original data. The sample standard deviation is also sensitive to extreme values ​​(outliers) in the data, as well as the sample variance.

### Function created to calculate the standard deviation

```python
def desvio_padrao(lista):
    n = len(lista) # number of list elements
    media = sum(lista) / n # average of the list
    soma_quadrados = sum((x - media) ** 2 for x in lista)
    variancia_amostral = soma_quadrados / (n - 1)
    desvio_padrao = variancia_amostral ** 0.5 # square root calculation
    return desvio_padrao
```

## Total Amplitude
- is a statistical measure that quantifies the total variation between the minimum and maximum values ​​of a data set. It is the difference between the maximum value and the minimum value of the sample
- The formula for calculating the total amplitude is as follows:
  
  Total Amplitude = Maximum Value - Minimum Value


### Function created to calculate the Total Amplitude

```python
def amplitude__total(lista):
    valor_minimo = min(lista) # valor máximo da lista
    valor_maximo = max(lista) # valor mínimo da lista
    amplitude_total = valor_maximo - valor_minimo
    return amplitude_total
```

## Q1 (first quartile)
- The first quartile, denoted Q1, is a statistical measure that divides a distribution of sorted data into four equal parts such that 25% of the data are below the first quartile and 75% of the data are above it

###Calculating the First Quartile involves the following steps:


* Sort the data in ascending order;
* Identify the position of the desired quartile, which is given by position = (n+1)/4, where n is the sample size;

- Check if the position is an integer or a decimal;
- If it is an integer, the first quartile is the value corresponding to that position in the sorted list;
- If it is a decimal, the first quartile is the average of the values ​​corresponding to the integer positions immediately below and above;

#### **In summary, the first quartile is the value that divides the sorted data into 25% below and 75% above**

### Function to calculate the First Quartile of a list of values

```python
def primeiro__quartil(lista):

    lista_ordenada = sorted(lista) # sorting the list
    n = len(lista_ordenada) # number of elements in ordered list
    indice_q1 = int(n * 0.25) # desired quartile index

    # get the value of the first quartile (Q1)
    q1 = lista_ordenada[indice_q1]
    return q1

```

##Q3 (third quartile)
- The third quartile, denoted as Q3, is a statistical measure that divides a distribution of sorted data into four equal parts, such that 75% of the data are below the third quartile and 25% of the data are above it;

As with the first quartile (Q1), calculating the third quartile involves the following steps:

- Sort data in ascending order;
- Identify the position of the desired quartile, which is given by position = (3n+1)/4, where n is the sample size;
- Check if the position is an integer or a decimal;
- If it is an integer, the third quartile is the value corresponding to that position in the sorted list;
- If it is a decimal, the third quartile is the average of the values ​​corresponding to the integer positions immediately below and above;

### Function Created to calculate the Third Quartile of a list of values


```python
def terceiro__quartil(lista):
    lista_ordenada = sorted(lista) # sorting the list
    n = len(lista_ordenada) # number of elements in ordered list
    indice_q3 = int(n * 0.75) # desired quartile index

    # obtain the value of the third quartile (Q3)
    q3 = lista_ordenada[indice_q3]
    return q3

```

##IQR (interquartile range)
- The interquartile range (IQR), also known as the interquartile range, is a statistical measure that represents the variation of data around the median. It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1) of a data distribution;

The formula for calculating the interquartile range is:

- IQR = Q3 - Q1

IQR is a robust measure that disregards extreme values ​​(outliers) and focuses on data variation within the central range of the distribution. It provides an estimate of the dispersion of the data around the median, being less sensitive to extreme values ​​compared to the total range;

### Function to calculate the IQR of a list of values

```python
def aiq_lista(lista):
    q1 = primeiro__quartil(lista)
    q3 = terceiro__quartil(lista)
    aiq = q3 - q1
    return aiq
```

##Lower Limit (Li from boxplot)
- The lower limit (Li) in a boxplot is a value that defines the lower limit to identify possible outliers (extreme values) in the data distribution. It is a statistical measure used to analyze asymmetry and the presence of outliers in the distribution.

The lower bound calculation can be performed using the interquartile range (IQR) and the 1.5 times IQR rule. The formula for calculating the lower bound is:
- Li = (Q1 - 1.5) * IQR

### Function to calculate Li

```python
def limite_inferior(lista):
    q1 = primeiro__quartil(lista)
    aiq = aiq_lista(lista)
    li = q1 - 1.5 * aiq
    return li
```

##Upper Limit (Ls of the boxplot)
- The upper limit (Ls) in a boxplot is a value that defines the upper limit to identify possible outliers (extreme values) in the data distribution. Like the lower limit (Li), the upper limit is a statistical measure used to analyze asymmetry and the presence of outliers in the distribution.

The upper bound calculation can be performed using the interquartile range (IQR) and the 1.5 times IQR rule. The formula for calculating the upper limit is:

- Ls = (Q3 + 1.5) * IQR

### Function to calculate the Ls of a list of values

```python
def limite_superior(lista):
    q3 = terceiro__quartil(lista)
    aiq = aiq_lista(lista)
    ls = q3 + 1.5 * aiq
    return ls
```

### Function to identify possible outliers and classify them

```python
from rich.console import Console
console=Console()
def possiveis_outliers(lista):
    # sort list in ascending order
    lista_ordenada = sorted(lista)

    # Calculation of list size and quartile index
    tamanho_lista = len(lista_ordenada)
    indice_q1 = int(tamanho_lista * 0.25) #Index of the first quartile
    indice_q3 = int(tamanho_lista * 0.75) # Index of the third quartile
    q1 = lista_ordenada[indice_q1]
    q3 = lista_ordenada[indice_q3]
    aiq = q3 - q1 # Interquartile range
    li = q1 - 1.5 * aiq # Lower bound for outliers
    ls = q3 + 1.5 * aiq # Upper bound for outliers

    # Find outliers below the lower bound (li) and above the upper bound (ls)
    outliers_baixos = [num for num in lista if num < li]
    outliers_altos = [num for num in lista if num > ls]

    # Check if outliers were found and display the result
    if len(outliers_baixos) > 0:
        return console.print(f'List has outliers! Low outliers: {outliers_baixos}')
    if len(outliers_altos) > 0:
        return console.print(f'List has outliers! High Outliers: {outliers_altos}')
    else:
      return console.print('List has no outliers!')
```

##Pearson's Correlation Coefficient(r)

####How can we interpret the values ​​that ρ can take?

- ρ = 0.7 to 09 (positive or negative): strong correlation;
- ρ = 0.5 to 0.7 (positive or negative): moderate correlation;
- ρ = 0.3 to 0.5 (positive or negative): weak correlation;
- ρ = 0 to 0.3 (positive or negative): no correlation.

### Function created to calculate the Pearson(r) Correlation

```python
from rich.console import Console
console=Console()
def correlacao_pearson(lista1, lista2):
      n = len(lista1)

      # Calculates the sums of variables
      soma_x = sum(lista1)
      soma_y = sum(lista2)

      # Calculates the sums of squares of the variables
      soma_quadrados_x = sum([x ** 2 for x in lista1])
      soma_quadrados_y = sum([y ** 2 for y in lista2])

      # Calculates the sum of the products of the variables
      soma_produtos = sum([x * y for x, y in zip(lista1, lista2)])

      # Calculates the numerator and denominator of the correlation coefficient
      numerador = n * soma_produtos - soma_x * soma_y
      denominador = ((n * soma_quadrados_x - soma_x ** 2) * (n * soma_quadrados_y - soma_y ** 2)) ** 0.5

      # Calculates Pearson's correlation coefficient
      r = numerador / denominador

      return console.print(f"Pearson Correlation Coefficient: {r:>37.2f}")
```

##Coefficient of Determination (R²)
- The coefficient of determination, also known as R², is a statistical measure that indicates the proportion of the variability of the dependent variable that can be explained by the independent variable in a regression model. In other words, it represents the amount of variation in the dependent variable that is explained by the linear regression model.

### The coefficient of determination varies between 0 and 1, where:

- A value close to 0 indicates that the regression model is not able to explain the variation in the dependent variable;
A value close to 1 indicates that the regression model is able to explain most of the variation in the dependent variable;

###Understanding the calculation of R²
The R², also called the Coefficient of Determination, tells how well my model is predicting correctly. Its calculation involves three measurements:
- Total Sum of Squares (STQ): shows the variation of y around the mean itself. It is the sum of the differences between the actual target value and its average squared;

**y** is the real value, **y with hat** is the mean

- Sum of Squared Residuals (SQU): variation of Y that is not explained by the elaborated model. It is the sum of the differences between the predicted value and the actual value squared;

r22.png

**y** is the actual value, **y with hat** is the predicted one

- Regression Sum of Squares (SQR): offers the variation of Y considering the X variables used in the model. It is the difference between the value of SQT and SQU. Therefore: SQR = SQT — SQU
The value of R² is the SQR itself divided by the SQT, that is, it is the division of the variation explained by the total variation of the data;

R2_final.png

###Function to calculate the Coefficient of Determination (R²)

```python
from rich.console import Console
console=Console()
def coeficiente_determinacao(lista_x, lista_y):

    # Calculate the number of data points
    n = len(lista_x)

    # Calculate the sum of x values and y values
    soma_x = sum(lista_x)
    soma_y = sum(lista_y)

    # Calculate the sum of products of x and y
    soma_xy = sum([x * y for x, y in zip(lista_x, lista_y)])

    # Calculate the sum of squared x values and y values
    soma_x_quadrado = sum([x ** 2 for x in lista_x])
    soma_y_quadrado = sum([y ** 2 for y in lista_y])

    # Calculate the numerator and denominator for the coefficient of determination
    numerador = (n * soma_xy) - (soma_x * soma_y)
    denominador = ((n * soma_x_quadrado) - (soma_x ** 2)) * ((n * soma_y_quadrado) - (soma_y ** 2))

    # Calculate the coefficient of determination (R²)
    coef_determinacao = numerador ** 2 / denominador
    coef_determinacao = coef_determinacao * 100

    # Print the coefficient of determination (R²)
    return console.print(f"Coefficient of determination R²: {coef_determinacao:>38.2f} %")
```

## Coefficient α or "alpha"
- The α or "alpha" coefficient of the simple linear regression equation is known as the slope coefficient or slope coefficient of the regression line;

- The intercept α or constant of the model, represents the expected value of Y when the controlled value of X is null;

### In a simple linear regression, the general equation is given by:
- y = α + βx

### Where:
- y is the dependent variable (or response) that we wish to predict;
- x is the independent (or predictor) variable we use to predict y;
- α is the alpha coefficient, which represents the intersection point of the line with the y axis (value of y when x = 0);
- β is the beta coefficient, which represents the slope (rate of change) of the line;

In the simple linear regression equation, the alpha coefficient is estimated based on the sample values ​​and is used along with the beta coefficient to make predictions or analyze the relationship between the x and y variables;


##Coeficiente β ou "beta"
- O coeficiente β da equação de regressão linear simples é o coeficiente angular da reta de regressão. Ele representa a mudança média esperada na variável dependente (Y) para cada unidade de mudança na variável independente (X).
- O coeficiente β ou "beta" é calculado utilizando a seguinte fórmula:

### Function created to calculate the coefficient α and β

```python
from rich.console import Console
console = Console()

def alpha_beta(lista_x, lista_y):
    n = len(lista_x)  # Get the length of the input lists
    somatorio_x = 0  # Variable to store the sum of elements in lista_x
    somatorio_y = 0  # Variable to store the sum of elements in lista_y
    somatorio_xy = 0  # Variable to store the sum of products of corresponding elements in lista_x and lista_y
    somatorio_x2 = 0  # Variable to store the sum of squared elements in lista_x

    # Calculate the sums needed for the linear regression formula
    for i in range(n):
        x = lista_x[i]  # Get the current element from lista_x
        y = lista_y[i]  # Get the current element from lista_y
        somatorio_x += x  # Add the current element to somatorio_x
        somatorio_y += y  # Add the current element to somatorio_y
        somatorio_xy += x * y  # Add the product of current elements to somatorio_xy
        somatorio_x2 += x ** 2  # Add the squared current element to somatorio_x2

    # Calculate the coefficients alpha and beta for the linear regression equation

    beta = (n * somatorio_xy - somatorio_x * somatorio_y) / (n * somatorio_x2 - somatorio_x ** 2)
    alpha = (somatorio_y - beta * somatorio_x) / n


    # Print the calculated coefficiente
    console.print(f"Alpha coefficient: {alpha:>51.2f}")
    console.print(f"Beta coefficient: {beta:>52.2f}")
```

### Function created, with error handling, where the user can only inform the script of numerical sequences separated by spaces

```python
def sequencia_numeros():
    while True:
        # Prompt the user to input a sequence of numbers separated by spaces
        sequencia = input("Enter a sequence of numbers separated by spaces: ")

        # Split the input into individual elements
        elementos = sequencia.split()

        # Check if all elements are numeric
        if not all(num.replace('.', '').replace('-', '').isdigit() for num in elementos):
            # Print an error message if the sequence contains non-numeric characters
            print("The sequence contains non-numeric characters. Please try again.")
            continue

        # Convert the elements into a list of floating-point numbers
        lista_numeros = [float(num) for num in elementos]

        # Return the list of numbers
        return lista_numeros
```
